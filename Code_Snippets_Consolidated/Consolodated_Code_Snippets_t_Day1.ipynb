```python
from newspaper import Article
import schedule
import time
import json
from google.colab import drive
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup


```

```python
# Mount your Google Drive
drive.mount('/content/drive')
```

```python
news_sources = [
    {
        'url': 'https://www.thesun.co.uk/news/worldnews/',
        'links_selector': 'a[href*="/news/"][href*="-"]'
        
    },
{
        'url': 'https://www.theguardian.com/world',
        'links_selector': 'a.u-faux-block-link__overlay.js-headline-text'
    },
    {
        'url': 'https://www.nytimes.com/section/world',
        'links_selector': 'div.css-1l4spti a'
    }
]
```

```python
def get_article_links(url, links_selector):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    article_links = [urljoin(url, a['href']) for a in soup.select(links_selector)]
    return article_links
    
def scrape_article(url):
    article = Article(url)
    article.download()
    article.parse()

    return {
        'title': article.title,
        'author': ', '.join(article.authors) if article.authors else '',
        'date': article.publish_date.strftime('%Y-%m-%d %H:%M:%S') if article.publish_date else None,
        'content': article.text,
        'source': article.source_url
    }

def job():
    print("Scraping news websites...")
    all_articles = []

    for source in news_sources:
        main_page_url = source['url']
        links_selector = source['links_selector']
        article_links = get_article_links(main_page_url, links_selector)

        for article_url in article_links:
            print(f"Scraping URL: {article_url}")
            article = scrape_article(article_url)
            all_articles.append(article)

            # Print article details
            print("Title:", article['title'])
            print("Author:", article['author'])
            print("Published Date:", article['date'])
            print("Content:", article['content'])
            print("\n")

    # Save articles to a JSON file in your Google Drive
    file_path = '/content/drive/MyDrive/scraped_articles.json'
    with open(file_path, 'w') as f:
        json.dump(all_articles, f, ensure_ascii=False, indent=4)

    print(f"Saved scraped articles to {file_path}\n")
```

```python
job()


while True:
    try:
        # Schedule the scraper to run every hour
        schedule.every(1).hours.do(job)

        # Run any pending scheduled jobs
        schedule.run_pending()

        # Wait for 1 second before checking the schedule again
        time.sleep(1)

    except Exception as e:
        print(f"Error occurred: {e}")
```
